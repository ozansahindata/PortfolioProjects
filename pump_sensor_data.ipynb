```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='ticks')
df = pd.read_csv('../DataBases/Public Industrial Datasets/Pump Sensor Data/sensor.csv')
```

```python
df.info()
# it looks like dataset has many null values in different columns
# Let's rename the columns first for a better readability 
```

```python
# Creating a dictionary 
sensor_dict = {"SENSOR_00" : "Motor Casing Vibration","SENSOR_01" : "Motor Frequency A","SENSOR_02" : "Motor Frequency B","SENSOR_03" : "Motor Frequency C","SENSOR_04" : "Motor Speed","SENSOR_05" : "Motor Current","SENSOR_06" : "Motor Active Power","SENSOR_07" : "Motor Apparent Power","SENSOR_08" : "Motor Reactive Power","SENSOR_09" : "Motor Shaft Power  ","SENSOR_10" : "Motor Phase Current A","SENSOR_11" : "Motor Phase Current B","SENSOR_12" : "Motor Phase Current C","SENSOR_13" : "Motor Coupling Vibration","SENSOR_14" : "Motor Phase Voltage AB","SENSOR_16" : "Motor Phase Voltage BC","SENSOR_17" : "Motor Phase Voltage CA","SENSOR_18" : "Pump Casing Vibration","SENSOR_19" : "Pump Stage 1 Impeller Speed","SENSOR_20" : "Pump Stage 1 Impeller Speed","SENSOR_21" : "Pump Stage 1 Impeller Speed","SENSOR_22" : "Pump Stage 1 Impeller Speed","SENSOR_23" : "Pump Stage 1 Impeller Speed","SENSOR_24" : "Pump Stage 1 Impeller Speed","SENSOR_25" : "Pump Stage 2 Impeller Speed","SENSOR_26" : "Pump Stage 2 Impeller Speed","SENSOR_27" : "Pump Stage 2 Impeller Speed","SENSOR_28" : "Pump Stage 2 Impeller Speed","SENSOR_29" : "Pump Stage 2 Impeller Speed","SENSOR_30" : "Pump Stage 2 Impeller Speed","SENSOR_31" : "Pump Stage 2 Impeller Speed","SENSOR_32" : "Pump Stage 2 Impeller Speed","SENSOR_33" : "Pump Stage 2 Impeller Speed","SENSOR_34" : "Pump Inlet Flow","SENSOR_35" : "Pump Discharge Flow","SENSOR_36" : "Pump UNKNOWN","SENSOR_37" : "Pump Lube Oil Overhead Reservoir Level","SENSOR_38" : "Pump Lube Oil Return Temp","SENSOR_39" : "Pump Lube Oil Supply Temp","SENSOR_40" : "Pump Thrust Bearing Active Temp","SENSOR_41" : "Motor Non Drive End Radial Bearing Temp 1","SENSOR_42" : "Motor Non Drive End Radial Bearing Temp 2","SENSOR_43" : "Pump Thrust Bearing Inactive Temp","SENSOR_44" : "Pump Drive End Radial Bearing Temp 1","SENSOR_45" : "Pump non Drive End Radial Bearing Temp 1","SENSOR_46" : "Pump Non Drive End Radial Bearing Temp 2","SENSOR_47" : "Pump Drive End Radial Bearing Temp 2","SENSOR_48" : "Pump Inlet Pressure","SENSOR_49" : "Pump Temp Unknown","SENSOR_50" : "Pump Discharge Pressure 1","SENSOR_51" : "Pump Discharge Pressure 2"}
# Just renaming columns starting with "sensor" keyword and excluding sensor_15 because there is no value in this col.
for col in df.columns:
    
    if "sensor" in col and col != "sensor_15":
        df[sensor_dict[col.upper()]] = df[col]

df['No'] = df['Unnamed: 0']
# Dropping the old and unnecessary columns
df = df.drop(df.filter(regex='sensor').columns, axis=1)
df = df.drop(['Pump UNKNOWN','Unnamed: 0'],axis=1)
```

```python
df.describe()
```

```python
df.head(10)
# Ok, it looks better with new columns
```

```python
# Lets drop the missing data
df.dropna(inplace=True)
df.shape
```

```python
# drop duplicates if they exist
df.drop_duplicates(inplace=True)
df.shape
```

```python
# There is a column called machine_status which has three values "NORMAL",'RECOVERING' and 'BROKEN'
# Filtering out just "NORMAL" condition will give more clarity while exploring the dataset.
df = df[df['machine_status']=='NORMAL']
df.shape
```

```python
# Ok lets fast check the correlations and see which columns are related.

df_numeric = df.select_dtypes('number') # Just selecting the numeric values
```

```python
sns.heatmap(df_numeric.corr())
plt.show()
plt.rcParams['figure.figsize'] = [18, 12]
plt.rcParams['figure.dpi'] = 200
%matplotlib inline
```

```python
# We can check out the vibrations first since they may effect pump working efficiency and cause system failure.
print(df_numeric.corr()['Motor Casing Vibration'].sort_values(ascending=False).head(10))
```

```python
# Motor casing vibration is correlated with motor speed and motor current, which makes sense when speed gets to it's
# working nominal values, vibration should remain the same in those speeds
# Let's check it out
sns.set_theme(style="darkgrid")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True) # just some make up
sns.relplot(data=df,x='Motor Casing Vibration',y='Motor Speed',hue='Motor Current',size='Motor Active Power',palette=cmap)
plt.grid(True)

plt.rcParams['figure.dpi'] = 200
```

```python
# Motor vibration increases and varies between 1.5-2.5 mm when motor speed reaches 500 - 700 rpm operation range.
# The variation is necessary to regulate the water flow.
# We should better check out motor voltage values

sns.lineplot(data=df,x='Motor Casing Vibration',y='Motor Phase Voltage AB',color='red')
sns.lineplot(data=df,x='Motor Casing Vibration',y='Motor Phase Voltage BC',palette='blue')
sns.lineplot(data=df,x='Motor Casing Vibration',y='Motor Phase Voltage CA',palette='green')
plt.grid(True)
plt.rcParams['figure.dpi'] = 200
```

```python
# Normal operating range is between 400-500 V for each three phases. The ones that vary between ~ 120 - 550 are
# probably tripping conditions. This is why vibration reaches it's peak.
# Motor Speed is more related with the subject of motor casing vibration.

#Let's check Coupling Vibrations.
```

```python
print(df_numeric.corr()['Motor Coupling Vibration'].sort_values(ascending=False).head(10))
```

```python
sns.relplot(data=df,x='Motor Coupling Vibration',y='Motor Speed',hue='Motor Casing Vibration',palette=cmap)
plt.grid(True)
```

```python
# Motor Coupling Vibration is also increses in normal speed operation range.
# Let's check the Pump Casing Vibration
```

```python
print(df_numeric.corr()['Pump Casing Vibration'].sort_values(ascending=False).head(10))
```

```python
# hmmm it's highly correlated with Voltage data. Without a graph I can say that Change in the voltage is changing
# the impellar speed  and creating a pump casing vibration.
```

```python
sns.relplot(data=df,x='Pump Casing Vibration',y='Motor Speed',hue='Motor Phase Voltage AB',palette=cmap)
plt.grid(True)
```

```python
# What this T-shape graphic tells us is that, acceptable vibration range is between 2-3mm if pump motor operates in 
# a stable 400-480 V range. When voltage decreases below 400 while operating in normal speed, vibration increases.
sns.relplot(data=df,x='Motor Current',y='Motor Phase Voltage AB',hue='Pump Casing Vibration',palette=cmap)
plt.grid(True)
```

```python
# My final assumption is pump casing vibration and motor coupling vibration is directly getting effected 
# by voltage regulation, which is something necessary to change the amount of discharge flow if you don't have a well
# working regulating valve before pump suction. To eliminiate Voltage regulations my advice is;

# 1 - A fluid coupling between motor and pump can be installed
# 2 - Regulating valves should get checked if they are working properly
# 3 - Recirculating valve should be checked if it's working properly,While testing and commissioning the pump.

# With this amount of data instead of prediction of failure, prevention of it is possible. 
```
